# Investment strategy using python 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Create lists
Year = []
Yearly_Income = []
Yearly_Expense = []
Annual_Returns = []
```

```python
# Assuming an yearly income
income = 35000
# Assuming an yearly expense
expense = income / 2
#Let's imagine in investing in an ETF that replicates the SP500 that has an 8% return 
# Average interest
interest_rate = 0.08
# You invest half of your income
investment = income / 2
# Calculating annual return
annual_return = investment*interest_rate
#Getting current year
year = 2021
```

```python
Yearly_investment = [] 
```

```python
# Append the first values to the lists 
Year.append(year)
Yearly_Income.append(income)
Yearly_Expense.append(expense)
Yearly_investment.append(investment)
Annual_Returns.append(annual_return)
```

```python
# Loop for 'n' years
invested_years = 20 #doing this for 20 years
for i in range(0,invested_years-1):
    # Update the investment to be the previous investment plus the previous annual return plus half of the income 
    investment = investment + annual_return + income/2
    #Update the annual return to be the current investment times the interest rate
    annual_return = investment * interest_rate
    Year.append(year+i+1)
    Yearly_Income.append(income)
    Yearly_Expense.append(expense)
    Yearly_investment.append(investment)
    Annual_Returns.append(annual_return)
```

```python
#Create a DataFrame
df = pd.DataFrame()
df['Year'] = Year
df['Yearly_Income'] = Yearly_Income
df['Yearly_Expenses'] = Yearly_Expense
df['Yearly_investment'] = Yearly_investment
df['Annual_Returns'] = Annual_Returns
```

```python
df
```

```python
# Visually show data
plt.figure(figsize=(12.2,4.5))
plt.plot(df['Year'], df['Yearly_Expenses'], label = 'Yearly_Expenses')
plt.plot(df['Year'], df['Annual_Returns'], label = 'Annual_Returns')
plt.xticks(rotation = 45)
plt.title('Retire in 10 Years')
plt.xlabel('Years')
plt.ylabel('$')
plt.legend()
plt.xticks(df['Year'])
plt.show()
```
